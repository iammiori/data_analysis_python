```python
import numpy as np
import pandas as pd
```

```python
df_movie = pd.read_csv('../source/movielens-20m-dataset/movie.csv')
df_movie.head()
# movie table 을 갖고온다
# movie table 에는 id, title, genres 가 있다.
```

```python
df_rating = pd.read_csv('../source/movielens-20m-dataset/rating.csv')
df_rating.head()
# ratin table 을 갖고온다
# rating table 에는 userid, movie id, ration, timestamp 가 있다
```

```python
df_movie = df_movie.iloc[:,:-1]
# genres 는 제외한다. 어차피 평점으로 추천을 해주기 때문에
#df_movie.head()
df_rating = df_rating.iloc[:,:-1]
# timestamp 도 굳이 필요하지 않은 정보이기 때문에 제외
```

movie table 과 rating table을 movied id로 merge

```python
df_total = pd.merge(df_movie,df_rating)
df_total.head(10)
#20000263 rows × 4 columns
```

```python
df_total = df_total.iloc[:1000000,:]
# data가 너무 많기도 하고, 전체 데이터를 넣으면 jupyter 자체에서
# memory 오류가 뜬다. 따라서 개수 제한
```

```python
pivot_table = df_total.pivot_table(index = ["userId"],columns = ["title"],values = "rating")
# pivot 은 내가 원하는 데이터로 , 데이터를 배치하여 새로운 dataframe을 만드는 것이다.

#pivot_final = pivot_table.fillna(0)
# nan 값을 0으로 채워주게 되면, 마치 user가 0으로 평가한것 처럼 
# 컴퓨터가 해석하기 때문에, nan 값 그대로 둔다.

pivot_table.head(20)
#pivot_final.to_csv('tmp/pivot_final.csv')
#132027 rows × 294 columns
```

```python
watched = pivot_table['Vampire in Brooklyn (1995)']
similarity = pivot_table.corrwith(watched)

# pandas 를 통해 Pearson Correlation Coefficient 를 구할수 있음
# 특히, corrwith() 는 모든 변수간 상관계수를 계산해줌.

similarity_order = similarity.sort_values(ascending=False)
# 내림차순으로 정렬해서, 추천해줄 영화 쉽게 구하기

similarity_order.head()
```
